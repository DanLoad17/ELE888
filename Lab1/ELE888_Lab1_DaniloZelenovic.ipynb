```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Load the dataset
url = 'https://raw.githubusercontent.com/yoga-suhas-km/Intelligent_systems/main/marks_dataset.csv'
df = pd.read_csv(url)

# Extracting midterm and final marks as arrays
markMT = df['Midterm mark'].values
markFinal = df['Final mark'].values

# Function to standardize features
def standardize_feature(feature):
    mean = np.mean(feature)
    std_dev = np.std(feature)
    return (feature - mean) / std_dev

# Function to calculate cost function (error)
def calculate_error(m, b, x, y):
    N = len(x)
    return np.sum((y - (m * x + b))**2) / N

# Function to compute gradient partial derivatives
def compute_gradients(m, b, x, y):
    N = len(x)
    dm = -2/N * np.sum(x * (y - (m * x + b)))
    db = -2/N * np.sum(y - (m * x + b))
    return dm, db

# Function to update m and b using learning rate alpha
def update_parameters(m, b, dm, db, alpha):
    m_new = m - alpha * dm
    b_new = b - alpha * db
    return m_new, b_new

# Initialize values
m = -0.5
b = 0
alpha = 0.0001

# Plot initial data without standardization
plt.scatter(markMT, markFinal, label='Data points')
plt.plot(markMT, m * markMT + b, color='red', label='Initial Regression Line')
plt.xlabel('Midterm Mark')
plt.ylabel('Final Mark')
plt.legend()
plt.title('Initial Data and Regression Line (Without Standardization)')
plt.show()

# Update parameters and plot without standardization for 100 iterations
for iteration in range(100):
    dm, db = compute_gradients(m, b, markMT, markFinal)
    m, b = update_parameters(m, b, dm, db, alpha)

# Plot the updated regression line and data points without standardization
plt.scatter(markMT, markFinal, label='Data points')
plt.plot(markMT, m * markMT + b, color='blue', label='Updated Regression Line (100 iterations)')
plt.xlabel('Midterm Mark')
plt.ylabel('Final Mark')
plt.legend()
plt.title('Updated Regression Line and Data Points (Without Standardization, 100 iterations)')
plt.show()

# Plot error without standardization for the first 100 iterations
errorVal = [calculate_error(m, b, markMT, markFinal)]
for iteration in range(100):
    dm, db = compute_gradients(m, b, markMT, markFinal)
    m, b = update_parameters(m, b, dm, db, alpha)
    errorVal.append(calculate_error(m, b, markMT, markFinal))

plt.plot(range(101), errorVal)
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error at each iteration (Without Standardization, 100 iterations)')
plt.show()

# Standardize features using the correct function
markMT_std = standardize_feature(markMT)
markFinal_std = standardize_feature(markFinal)

# Reset parameters for standardized features
m = -0.5
b = 0

# Plot initial data with standardization
plt.scatter(markMT_std, markFinal_std, label='Data points')
plt.plot(markMT_std, m * markMT_std + b, color='red', label='Initial Regression Line (With Standardization)')
plt.xlabel('Standardized Midterm Mark')
plt.ylabel('Standardized Final Mark')
plt.legend()
plt.title('Initial Data and Regression Line (With Standardization)')
plt.show()

# Update parameters and plot with standardized features for 100 iterations
for iteration in range(100):
    dm, db = compute_gradients(m, b, markMT_std, markFinal_std)
    m, b = update_parameters(m, b, dm, db, alpha)

# Plot the updated regression line and data points with standardized features
plt.scatter(markMT_std, markFinal_std, label='Data points')
plt.plot(markMT_std, m * markMT_std + b, color='green', label='Updated Regression Line (100 iterations)')
plt.xlabel('Standardized Midterm Mark')
plt.ylabel('Standardized Final Mark')
plt.legend()
plt.title('Updated Regression Line and Data Points (With Standardization, 100 iterations)')
plt.show()

# Plot error with standardized features for the first 100 iterations
errorVal_std = [calculate_error(m, b, markMT_std, markFinal_std)]
for iteration in range(100):
    dm, db = compute_gradients(m, b, markMT_std, markFinal_std)
    m, b = update_parameters(m, b, dm, db, alpha)
    errorVal_std.append(calculate_error(m, b, markMT_std, markFinal_std))

plt.plot(range(101), errorVal_std)
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error at each iteration (With Standardization, 100 iterations)')
plt.show()

# Reset parameters for 2000 iterations without standardization
m = -0.5
b = 0

# Update parameters and plot without standardization for 2000 iterations
for iteration in range(2000):
    dm, db = compute_gradients(m, b, markMT, markFinal)
    m, b = update_parameters(m, b, dm, db, alpha)

# Plot the updated regression line and data points without standardization
plt.scatter(markMT, markFinal, label='Data points')
plt.plot(markMT, m * markMT + b, color='purple', label='Updated Regression Line (2000 iterations)')
plt.xlabel('Midterm Mark')
plt.ylabel('Final Mark')
plt.legend()
plt.title('Updated Regression Line and Data Points (Without Standardization, 2000 iterations)')
plt.show()

# Plot error without standardization for the first 2000 iterations
errorVal = [calculate_error(m, b, markMT, markFinal)]
for iteration in range(2000):
    dm, db = compute_gradients(m, b, markMT, markFinal)
    m, b = update_parameters(m, b, dm, db, alpha)
    errorVal.append(calculate_error(m, b, markMT, markFinal))

plt.plot(range(2001), errorVal)
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error at each iteration (Without Standardization, 2000 iterations)')
plt.show()

# Reset parameters for 2000 iterations with standardization
m = -0.5
b = 0

# Update parameters and plot with standardized features for 2000 iterations
for iteration in range(2000):
    dm, db = compute_gradients(m, b, markMT_std, markFinal_std)
    m, b = update_parameters(m, b, dm, db, alpha)

# Plot the updated regression line and data points with standardized features
plt.scatter(markMT_std, markFinal_std, label='Data points')
plt.plot(markMT_std, m * markMT_std + b, color='orange', label='Updated Regression Line (2000 iterations)')
plt.xlabel('Standardized Midterm Mark')
plt.ylabel('Standardized Final Mark')
plt.legend()
plt.title('Updated Regression Line and Data Points (With Standardization, 2000 iterations)')
plt.show()

# Plot error with standardized features for the first 2000 iterations
errorVal_std = [calculate_error(m, b, markMT_std, markFinal_std)]
for iteration in range(2000):
    dm, db = compute_gradients(m, b, markMT_std, markFinal_std)
    m, b = update_parameters(m, b, dm, db, alpha)
    errorVal_std.append(calculate_error(m, b, markMT_std, markFinal_std))

plt.plot(range(2001), errorVal_std)
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error at each iteration (With Standardization, 2000 iterations)')
plt.show()

# Verify the results using scikit-learn
model = LinearRegression()

# Fit the model without standardization
model.fit(markMT.reshape(-1, 1), markFinal)
markFinal_pred = model.predict(markMT.reshape(-1, 1))

# Print coefficients and intercept without standardization
print(f'Coefficients without standardization: {model.coef_[0]}')
print(f'Intercept without standardization: {model.intercept_}')

# Fit the model with standardized features
scaler = StandardScaler()
markMT_std_sklearn = scaler.fit_transform(markMT.reshape(-1, 1))
markFinal_std_sklearn = scaler.fit_transform(markFinal.reshape(-1, 1))

model.fit(markMT_std_sklearn, markFinal_std_sklearn)
markFinal_pred_std = model.predict(markMT_std_sklearn)

# Print coefficients and intercept with standardization
coefficients_with_std = model.coef_[0] / np.std(markMT)
intercept_with_std = model.intercept_ - np.sum(coefficients_with_std * np.mean(markMT) / np.std(markMT))
print(f'Coefficients with standardization: {coefficients_with_std}')
print(f'Intercept with standardization: {intercept_with_std}')
# Verify the results using scikit-learn
model = LinearRegression()

# Fit the model without standardization
model.fit(markMT.reshape(-1, 1), markFinal)
markFinal_pred = model.predict(markMT.reshape(-1, 1))

# Print coefficients and intercept without standardization
print(f'Coefficients without standardization: {model.coef_[0]}')
print(f'Intercept without standardization: {model.intercept_}')

# Fit the model with standardized features
scaler = StandardScaler()
markMT_std_sklearn = scaler.fit_transform(markMT.reshape(-1, 1))
markFinal_std_sklearn = scaler.fit_transform(markFinal.reshape(-1, 1))

model.fit(markMT_std_sklearn, markFinal_std_sklearn)
markFinal_pred_std = model.predict(markMT_std_sklearn)

# Print coefficients and intercept with standardization
coefficients_with_std = model.coef_[0] / np.std(markMT)
intercept_with_std = model.intercept_ - np.sum(coefficients_with_std * np.mean(markMT) / np.std(markMT))
print(f'Coefficients with standardization: {coefficients_with_std}')
print(f'Intercept with standardization: {intercept_with_std}')
```
